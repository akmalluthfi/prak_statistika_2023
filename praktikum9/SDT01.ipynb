```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style = 'whitegrid')
plt.figure(figsize = (5,8))
ax = sns.boxplot(y ='Credit_Score',data = data_csv,orient = 'y',color = 'orange')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style = 'whitegrid')
plt.figure(figsize = (10,8))
ax = sns.boxplot(y ='Loan_Amt',data = data_csv,orient = 'v',color = 'green')
```

```python
type_counts = data_csv['Credit_Risk'].value_counts()
df = pd.DataFrame({'Risiko Kredit':type_counts},
                      index =['Moderate','High','Low','Very Low'])
df.plot.pie(y='Risiko Kredit',figsize = (6,6),autopct='%1.2f%%')
```

```python
plt.figure(figsize=(10,5))
sns.distplot(data_csv['Credit_Score'],kde=True)
```

```python
filter_data=data_csv.dropna(subset=['Credit_Score'])
plt.figure(figsize=(10,5))
sns.distplot(filter_data['Credit_Score'],kde=True)
```

```python
sns.set(style='darkgrid')
plt.figure(figsize = (20,10))
ax=sns.countplot(x='Credit_Risk',data=data_csv)
```

```python
df=data_csv.groupby(['Credit_Risk'])['Late_Payments'].sum().reset_index()
df['Late_Payments'].plot(kind='hist')
```
